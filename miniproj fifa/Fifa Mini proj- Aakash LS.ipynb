# Converting height into cm

```python
Height_cm = []
for i in list(df['Height'].values):
    try:
        Height_cm.append((float(str(i)[0])*12.0 + float(str(i)[2:]))*2.54)
    except(ValueError):
        Height_cm.append(np.nan)
        
df['Height_cm'] = Height_cm
```

## 3. Avg. Potential on all positions(Potential vs position)

```python
positions = []
for i in df['Position'].unique():
    positions.append(i)
for j in positions:
    temp = df.loc[df['Position']==j]
    plt.title('Average potential on all positions')
    bars =plt.bar(j,np.average(temp['Overall']))
```

## Conclusion
### It can be concluded that the most common position out of the given players is LF followed by RF (Although all others do not vary much)


## 4. Reactions vs overall


```python
x=df['Reactions']
y=df['Overall']
plt.figure(figsize=(15,8))
plt.bar(x,y)
plt.title("Reaction Vs overall",fontsize=20)
plt.xlabel("Reactions")
plt.ylabel("Overall Rating")

plt.show()
```

```python
#Unique values of reactions
df['Reactions'].unique()
```

## Conclusion
### From the graph , it is clearly visible that , more the reactions of the player, more is their overall rating!
